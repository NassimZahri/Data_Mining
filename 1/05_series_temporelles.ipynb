# 05 — Séries Temporelles (Basique)
Indexation temporelle, resampling, moyennes mobiles, différences, et détection simple d'anomalies.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

DATA_DIR = Path('data')
ventes = pd.read_csv(DATA_DIR / 'ventes.csv', parse_dates=['date'])

ts = ventes.set_index('date').resample('D')['total'].sum().fillna(0)
ts.head()
```

## 1. Moyennes mobiles

```
plt.figure()
ts.plot(label='quotidien')
ts.rolling(7).mean().plot(label='MA 7j')
plt.title('CA quotidien et MA(7)')
plt.legend()
plt.show()
```

## 2. Différenciation (stationnariser grossièrement)

```
diff = ts.diff()
plt.figure()
diff.plot()
plt.title('Différence 1')
plt.show()
```

## 3. Détection simple d'anomalies (z-score)

```
mean = ts.mean()
std = ts.std()
z = (ts - mean) / std
anomalies = ts[np.abs(z) > 3]
anomalies.head()
```

```
plt.figure()
ts.plot()
plt.scatter(anomalies.index, anomalies.values)
plt.title('Anomalies (|z|>3)')
plt.show()
```
